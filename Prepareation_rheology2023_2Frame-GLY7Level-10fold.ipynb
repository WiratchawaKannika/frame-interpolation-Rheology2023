### ðŸ“‚ Prepare text. file to predict frame

```python
def prepare_text(fold, files_):
    pth2save = f'/media/SSD/Frame_Inter_rheology2023/_10GenFrame/ues2Frame/pred_text/GLY7Level-10fold/fold{fold}'
    if not os.path.exists(pth2save) :
        os.makedirs(pth2save)
    for k in range(len(files_)) :
        name_text = files_[k].split("/")[-1]
        pth = f'{pth2save}/{name_text}-2linedemo.txt'
        #print(pth)
        files_img = glob.glob(f"{files_[k]}/*")
        files_img.sort()
        df = pd.DataFrame(files_img, columns =['Path'])
        df_ = df[:-1]
        df_ = df_.reset_index(drop=True)
        df2 = pd.DataFrame(files_img, columns =['Path'])
        df2_ = df2[1:]
        df2_ = df2_.reset_index(drop=True)
        df_['Path_txt'] = ''
        for i in range(len(df_)):
            name1 = df_['Path'][i]
            name2 = df2_['Path'][i]
            df_.loc[df_.index[i], 'Path_txt'] = str(name1) + ' ' + str(name2) 
        list_path = df_['Path_txt'].tolist()
        #print(list_path)
        with open(f'{pth}', 'w') as f:
             for line in list_path:
                 f.write(f"{line}\n")
        print(f'[INFO] ({k+1}.) On Process : Write text file name -> [ {pth} ] ')   
        print("="*120)
```

## Function: Create Text Files

```python
import glob
import os

def Create_text(Pth_DataFrame, fold):
    DataFrame = pd.read_csv(Pth_DataFrame)
    set_fold = list(set(DataFrame["fold"]))
    set_fold.sort()
    print(f"Dataset 10-fold: {DataFrame.shape[0]} Folder")
    print(f"10-Fold ==> {set_fold}")

    DataFrame_fold = DataFrame[DataFrame["fold"]==fold].reset_index(drop=True)
    print(f"Dataset fold-{fold}: {DataFrame_fold.shape[0]} Folder")
    print("Fold ==> ", set(DataFrame_fold["fold"]))
    files_ = DataFrame_fold["Folder_pathimg"].tolist()
    print(len(files_))
    #print(files_)
    
    img_all = []
    for k in range(len(files_)):
        img_ = []
        img = glob.glob(f"{files_[k]}/*")
        #print(f"{k+1}. *Images in directory -> {files_[k]}--> [{len(img)}]")
        #print('-'*120)
        img_.append(len(img))
        img_all.append(img_)
        
    img_len = []
    for sublist in img_all:
        for item in sublist:
            img_len.append(item)
    print(len(img_len))
    print(f" à¸ˆà¸³à¸™à¸§à¸™à¸£à¸¹à¸›à¸ à¸²à¸žà¸—à¸±à¹‰à¸‡à¸«à¸¡à¸” {len(img_all)} Folder -> [{sum(img_len)}] à¸£à¸¹à¸›")
    
    ###*** set text file
    prepare_text(fold, files_)
    
    return
```
