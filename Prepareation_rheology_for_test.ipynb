````python
new_dir = '/media/SSD/rheology_data/Glycerol-water-mixtures-train/' 
print(new_dir)
dir_list = os.listdir(new_dir)
dir_list
````

````python
import glob

dir_save = '/media/SSD/rheology_data/'
dir_list = os.listdir(f'{dir_save}Glycerol-water-mixtures-train')
#print(len(dir_list))
dir_list.sort()
#dir_list
dir_sub_ = []
for i in range(len(dir_list)):
    dir_sub = sorted(glob.glob(f"{dir_save}Glycerol-water-mixtures-train/{dir_list[i]}/*"))
    dir_sub.sort()
    print(len(dir_sub))
    dir_sub_.append(dir_sub)
    print('-'*80)
    
for j in range(len(dir_sub_)):
    test_dir = []
    dir_sub = dir_sub_[j]
    #print(dir_sub)
    for d in dir_sub:
        #print(d)
        d_ = d.split('/')[5:]
        #print(d_)
        d1_ = '/'.join(d_)
        #print(d1_)
        test_dir.append(d1_)
    with open(f'{dir_save}testlist_f{j+1}.txt', 'w') as f:
             for line in test_dir:
                 f.write(f"{line}\n")
    print(f'On Process : Write text file name -> [ {dir_save}testlist_f{j+1}.txt ] ')   
````
